```python
df = pd.read_csv('Atlanta Dataset.csv', sep=',', error_bad_lines=False, index_col=False, dtype='unicode')
```

```python
df = df.dropna(how='all')
```

```python
df
```

```python
dfco = df[df['Specie'] == 'co']
```

```python
dfco
```

```python
dftemp = df[df["Specie"] == "temperature"]
```

```python
dftemp
```

```python
dfco = dfco.drop(['min', 'max', 'variance', 'Country', 'count'], axis = 1)
```

```python
dftemp = dftemp.drop(['min', 'max', 'variance', 'Country', 'count'], axis = 1)
```

```python
dfco
```

```python
dftemp
```

```python
dfco = dfco.sort_values(by = "Date")
```

```python
dftemp = dftemp.sort_values(by = "Date")
```

```python
dfco
```

```python
dftemp
```

```python
dfcotemp = pd.DataFrame(columns=['Date', 'City', 'Specie', 'median', 'medianTemp'])
```

```python
for date in dfco["Date"]:
  if date in dftemp["Date"].values:
    try:
        temp = float(dftemp.loc[dftemp['Date'] == date]['median'])
        dfcotemp = dfcotemp.append((dfco[dfco['Date'] == date]))
        dfcotemp.loc[dfcotemp['Date'] == date] = dfcotemp.loc[dfcotemp['Date'] == date].assign(medianTemp=temp)
        dfcotemp.loc[dfcotemp['Date'] == date] = dfcotemp.loc[dfcotemp['Date'] == date].assign(median=float(dfcotemp.loc[dfcotemp['Date'] == date]['median']))
    except TypeError:
        continue
```
